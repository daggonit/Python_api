### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display

import geoviews as gv
from geoviews import opts
from holoviews import dim
import geoviews.feature as gf
import cartopy.crs as ccrs

gv.extension('bokeh')

# Create points for each city
points = gv.Points(city_data_df, kdims=['Lng', 'Lat'], vdims=['City', 'Humidity'])

# Set options for points
points.opts(
    size=dim('Humidity')*0.1,  # Size by humidity
    tools=['hover'],
    color='red',
    width=800,
    height=600,
    projection=ccrs.PlateCarree()
)

# Create a map with the points
world_map = gf.coastline * points
world_map
```
